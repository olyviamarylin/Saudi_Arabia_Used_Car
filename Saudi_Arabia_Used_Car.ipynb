# **Saudi Arabia Used Car Price Prediction**

Pasar jual-beli mobil bekas di Arab Saudi berkembang pesat, terutama dengan meningkatnya permintaan pelanggan di berbagai wilayah. Salah satu platform jual-beli mobil bekas di Saudi Arabia berfungsi sebagai penghubung antara penjual dan pembeli. Platform ini menyediakan ruang bagi penjual untuk memasarkan mobil mereka dan memungkinkan calon pembeli untuk membandingkan berbagai pilihan berdasarkan tipe, merek, jarak tempuh, dan harga. Namun, dengan banyaknya daftar mobil yang tersedia, menetapkan harga yang optimal menjadi tantangan tersendiri. 

Dari dataset yang dimiliki dapat dianalisis yaitu terdapat 5624 catatan mobil bekas, terdapat informasi penting seperti merek, model, tahun pembuatan, asal, opsi kendaraan, kapasitas mesin, jenis transmisi, jarak tempuh (mileage), harga, dan negosiasi harga (Ya/Tidak). Analisis data ini penting untuk memahami tren harga mobil bekas, preferensi pelanggan, dan faktor-faktor yang memengaruhi harga pasar. Penjual harus dapat menentukan harga yang cukup kompetitif untuk menarik calon pembeli tanpa mengorbankan keuntungan mereka. Jika harga yang ditetapkan terlalu tinggi, pembeli cenderung akan berpaling, sementara harga yang terlalu rendah berisiko merugikan penjual. Proses penetapan harga mobil bekas melibatkan banyak variabel yang saling memengaruhi, sehingga memerlukan pendekatan yang lebih sistematis untuk efisiensi cost pada analisis Harga mobil bekas yang akan di jual.

**Permasalahan**
Platform tersebut dihadapkan pada tantangan besar dalam membantu penjual menetapkan harga yang tepat—harga yang tidak hanya menarik bagi calon pembeli tetapi juga memastikan keuntungan yang memadai bagi penjual. Banyak penjual sering kali menghadapi kesulitan dalam menentukan harga yang sesuai, meskipun mereka dapat membandingkan kendaraan serupa di platform tersebut. Penjual bebas menentukan harga jual mobil mereka. Meskipun mereka dapat menaikkan harga jika merasa mobil memiliki nilai tambah tertentu, persaingan yang ketat di antara penjual membuat strategi penetapan harga yang efektif menjadi sangat penting. Tanpa alat yang memadai, penjual mungkin kesulitan untuk menetapkan harga yang dapat bersaing di pasar.

**Tujuan**  
Berdasarkan permasalahan yang dimiliki membutuhkan sebuah alat analisis prediktif yang dapat membantu penjual menentukan harga jual mobil bekas yang lebih akurat. Fitur-fitur penting seperti merek, tahun produksi, jarak tempuh, ukuran mesin, dan lokasi penjualan harus dianalisis untuk mendapatkan estimasi harga yang lebih presisi. Dengan prediksi harga yang lebih akurat, penjual dapat menawarkan harga yang kompetitif, menarik lebih banyak pembeli, dan tetap memperoleh keuntungan yang optimal. Selain itu, keakuratan prediksi harga dapat meningkatkan tingkat keberhasilan transaksi di platform, yang pada akhirnya akan mendorong peningkatan volume penjualan, mempercepat pertumbuhan bisnis mobil bekas, dan memberikan kontribusi positif terhadap pendapatan platform melalui komisi penjualan.

**Algoritma/Permodelan**
Model yang digunakan adalah Regresi linear. Fungsi utama algoritma ini adalah menemukan parameter optimal yang meminimalkan error dalam prediksi. karna hubungan Linear Antar Variabel seperti Harga mobil bekas umumnya memiliki hubungan linear dengan faktor-faktor seperti tahun produksi, jarak tempuh (mileage), merek, kondisi mesin, dan fitur lainnya. Dalam dataset dengan pola yang relatif sederhana dan fitur yang saling berhubungan secara linear, regresi linear dapat memberikan performa yang konsisten.


**Metrik Evaluasi**  
Metrik evaluasi yang akan digunakan adalah RMSE (Root Mean Squared Error), MAE (Mean Absolute Error), dan MAPE (Mean Absolute Percentage Error).

- **RMSE (Root Mean Squared Error)**  
  RMSE adalah ukuran yang menghitung rata-rata akar dari kesalahan prediksi. Nilai RMSE yang lebih kecil menunjukkan bahwa model lebih akurat dalam memprediksi harga mobil. RMSE memberi penekanan lebih pada kesalahan besar, sehingga menjadi pilihan yang baik ketika kesalahan signifikan tidak dapat diterima. Oleh karena itu, RMSE efektif untuk memprediksi kesalahan besar yang dapat memiliki dampak signifikan pada hasil akhir.

- **MAE (Mean Absolute Error)**  
  MAE adalah ukuran yang menghitung rata-rata kesalahan absolut dari prediksi. Nilai MAE yang lebih kecil menunjukkan bahwa model lebih akurat dalam memprediksi harga mobil. Dengan memberikan bobot yang sama pada setiap kesalahan, MAE memberikan penilaian yang adil terhadap distribusi kesalahan. Selain itu, MAE berguna untuk meminimalkan dampak dari *outlier* dan kesalahan ekstrem, memungkinkan fokus pada kesalahan rata-rata yang lebih umum.

- **MAPE (Mean Absolute Percentage Error)**  
  MAPE adalah ukuran yang menghitung rata-rata persentase kesalahan prediksi. Nilai MAPE yang lebih kecil menunjukkan akurasi model yang lebih tinggi dalam memprediksi harga mobil. MAPE memberikan konteks yang lebih mudah dipahami, sehingga berguna untuk membandingkan kinerja model di berbagai dataset, terutama ketika ada variasi besar pada nilai aktual. Dengan menghitung kesalahan sebagai persentase dari nilai aktual, MAPE mengurangi pengaruh *outlier*, menjadikannya relevan untuk analisis bisnis di mana proporsi kesalahan penting. MAPE ideal untuk memahami kinerja model dalam konteks persentase, memastikan bahwa kesalahan yang lebih kecil relatif terhadap nilai yang lebih besar mendapatkan perhatian yang tepat.

**Handling Outliers**

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}') 
```

**Hyperparameter Tuning**

Hyperparameter SVR(Support Vector Regression) memiliki parameter yang mempengaruhi kinerja modelnya sebagai berikut:

- C: Dikontrol menggunakan distribusi uniform antara 0.1 hingga 100. Nilai C yang tinggi akan mencoba meminimalkan kesalahan training dengan keras, sedangkan nilai rendah akan membiarkan margin yang lebih luas untuk kesalahan kecil.

- epsilon: Dikontrol menggunakan distribusi uniform antara 0.01 hingga 0.5. epsilon menentukan margin di mana kesalahan prediksi tidak akan dihitung dalam fungsi loss SVR. Epsilon kecil jika presisi sangat penting, epsilon besar jika prediksi yang agak kasar dapat diterima.

- kernel: Dipilih dari opsi ['linear', 'rbf', 'poly']. Kernel menentukan fungsi pemetaan non-linear untuk memproyeksikan data ke dimensi yang lebih tinggi. Penggunaan *linear* jika data tampaknya bersifat linear, *rbf* jika pola data kompleks dan non-linear dan *poly* untuk hubungan non-linear dengan perilaku polinomial.

- gamma: Dipilih dari opsi ['scale', 'auto']. gamma menentukan seberapa jauh pengaruh satu titik pelatihan dapat menjangkau titik lain di ruang fitur. Nilai tinggi membuat pengaruh satu titik terbatas pada titik terdekat, sedangkan nilai rendah memperluas pengaruh tersebut. Nilai gamma tinggi (misalnya 1, 10), Nilai gamma rendah (misalnya 0.001). gamma tinggi untuk pola data yang sangat lokal dan kompleks.
gamma rendah jika pola data lebih global dan lebih umum.

```python
# Definisikan parameter SVR untuk tuning
hyperParamSVR = {
    'model__regressor__C': [0.1, 1, 10, 100],               # Regularization parameter
    'model__regressor__epsilon': [0.01, 0.1, 0.2, 0.5],     # Epsilon in the epsilon-SVR model
    'model__regressor__kernel': ['linear', 'rbf', 'poly'],  # Kernel type
    'model__regressor__gamma': ['scale', 'auto']            # Kernel coefficient
}

# Buat Pipeline untuk SVR
pipeSVR = Pipeline([
    ('Preprocess', transformer),             # Preprocessing data (transformer yang Anda gunakan sebelumnya)
    ('model', transform_svreg)               # Model SVR
])

# Lakukan RandomizedSearchCV untuk menemukan parameter terbaik
randomSearchSVR = RandomizedSearchCV(
    estimator=pipeSVR,
    param_distributions=hyperParamSVR,
    n_iter=50,                              # Jumlah iterasi untuk sampling parameter
    cv=5,                                   # Cross-validation sebanyak 5 lipatan
    scoring=['neg_mean_absolute_error',
             'neg_mean_absolute_percentage_error',
             'neg_mean_squared_error',
             'neg_root_mean_squared_error'],
    refit='neg_mean_absolute_percentage_error',  # Model dioptimalkan berdasarkan MAPE.
    random_state=37,
    verbose=2,                              # Menampilkan progres pencarian hyperparameter.
    n_jobs=-1
)

# Fitting model dengan data latih
randomSearchSVR.fit(Xtrain, ytrain)

# Tampilkan hasil terbaik
display(
    randomSearchSVR.best_estimator_,
    randomSearchSVR.best_params_,
    randomSearchSVR.best_score_,
    pd.DataFrame(randomSearchSVR.cv_results_).sort_values(['rank_test_neg_mean_absolute_percentage_error',
                                                           'rank_test_neg_mean_absolute_error']).head()
)
```



Hyperparameter XGBRegressor() memiliki parameter yang mempengaruhi kinerja modelnya sebagai berikut:

- max_depth : Mengontrol kedalaman maksimum dari setiap pohon keputusan, Kedalaman pohon memengaruhi seberapa kompleks aturan yang dapat dipelajari oleh model.

- learning_rate : Mengontrol ukuran langkah (step size) pada setiap iterasi boosting. Nilai kecil membuat model belajar secara perlahan, sedangkan nilai besar membuat model belajar lebih cepat.

- n_estimators : Menentukan jumlah pohon keputusan (decision trees) yang akan dibangun oleh model. Setiap pohon membantu memperbaiki kesalahan dari pohon sebelumnya dalam proses boosting.

```python
# Definisikan parameter XGB untuk tuning
hyperParamXGB = {'model__regressor__max_depth': list(np.arange(2, 10)), 
                 'model__regressor__learning_rate': list(np.arange(1, 101, 10)/100),
                 'model__regressor__n_estimators': list(np.arange(200, 301, 20))
}

# Buat Pipeline untuk XGB
pipeXGB = Pipeline([('Preprocess', transformer),
                    ('model',transform_xgbreg)])

# Lakukan RandomizedSearchCV untuk menemukan parameter terbaik
randomSearchXGB = RandomizedSearchCV(
    estimator=pipeXGB,
    param_distributions=hyperParamXGB,
    n_iter=50,                              # Jumlah iterasi untuk sampling parameter
    cv=5,                                   # Cross-validation sebanyak 5 lipatan
    scoring=['neg_mean_absolute_error',
             'neg_mean_absolute_percentage_error',
             'neg_mean_squared_error',
             'neg_root_mean_squared_error'],
    refit='neg_mean_absolute_percentage_error',
    random_state=37,
    verbose=2,
    n_jobs=-1
)

# Fitting model dengan data latih
randomSearchXGB.fit(Xtrain, ytrain)

# Tampilkan hasil terbaik
display(
    randomSearchXGB.best_estimator_,
    randomSearchXGB.best_params_,
    randomSearchXGB.best_score_,
    pd.DataFrame(randomSearchXGB.cv_results_).sort_values(['rank_test_neg_mean_absolute_percentage_error',
                                                           'rank_test_neg_mean_absolute_error']).head()
)
```
